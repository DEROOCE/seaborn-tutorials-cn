```python
# 解决numpy dtype warning problem
import warnings
warnings.filterwarnings("ignore", message="numpy.dtype size changed")
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# 显示分位数
sns.boxplot(x=tips['size'])
```

```python
# 两个变量的箱型图
sns.boxplot(x='sex', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='husl')
```

```python
sns.boxplot(x='day', y='total_bill', order=['Sat', 'Fri', 'Sun', 'Thur'], data=tips, palette='husl')
```

```python
# 设置boxplot的朝向
sns.boxplot(data=tips, orient='horizontal',  palette='coolwarm') # horizontal or h  vertical or v
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='husl')
# swarmplot ：每个点对应一个数据点，绘制数据的分布蜂窝图
sns.swarmplot(x='day', y='total_bill', data=tips, color='blue')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='husl')
# swarmplot ：每个点对应一个数据点，绘制数据的分布蜂窝图
sns.swarmplot(x='day', y='total_bill', data=tips, color='0.45')
```
