```python
# 解决numpy dtype warning problem
import warnings
warnings.filterwarnings("ignore", message="numpy.dtype size changed")
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")

import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid')

%matplotlib inline
```

```python
# 直方图绘制
num = np.random.randn(150)

sns.distplot(num, color='red')
```

distplot在绘制直方图的同时，会提供对数据适配的核密度估计(KDE)。

```python
# 联用pandas，为数据加入标签
label_dist = pd.Series(num, name='variable x')

sns.distplot(label_dist)
```

```python
# 将直方图横向观察
sns.distplot(label_dist, vertical=True)
```

```python
# 将直方图的bins隐藏，只显示分布曲线
sns.distplot(label_dist, hist=False)
```

```python
# 绘制rugplot（在每个观察点上的垂直小标签）
sns.distplot(label_dist, rug=True, hist=False)
```
